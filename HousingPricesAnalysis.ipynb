```python
housing=pd.read_csv("housing.csv")
```

```python
housing.head()
```

```python
housing.describe().T
```

```python
housing.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
fig=plt.figure(dpi=80, figsize=(6,4))
ax=fig.add_axes([1,1,1,1])
ax.set(xlabel="Median Income Class", ylabel="Frequency", title="Distribution of Median Income")
housing["median_income"].hist(bins=50, ax=ax)
plt.show()
```

```python
fig=plt.figure(dpi=80, figsize=(6,4))
ax=fig.add_axes([1,1,1,1])
ax.set(xlabel="Median Income Category", ylabel="Frequency", title="Distribution of Income Category")
housing["median_income"].hist(color = 'orange',ax=ax)
plt.show()
```

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
s=housing["population"]/100, label="population", figsize=(10,7),
c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```

```python
from pandas.plotting import scatter_matrix

attributes = ["median_house_value", "median_income", "total_rooms",
              "housing_median_age"]
scatter_matrix(housing[attributes], figsize=(12, 10))
plt.show()
```
